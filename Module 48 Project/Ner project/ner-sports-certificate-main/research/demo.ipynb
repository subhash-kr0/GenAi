```python
import spacy
import pandas as pd

# Load your spaCy NER model
nlp = spacy.load("D:/ner-sports-task/artifacts/11_11_2023_23_42_34/ModelTrainerArtifacts/model-best")
```

```python
# Load the new data from the Excel file
new_data_df = pd.read_excel('D:/ner-sports-task/data/data.xlsx', 'Data-Pending', header=None, names=['Certificate_Description'])
new_data_df
```

```python
# Extract information from the certificate descriptions

extracted_info = []
for desc in new_data_df['Certificate_Description']:
    doc = nlp(desc)
    sports_name = None
    winning_position = None
    org_year = None

    for ent in doc.ents:
        if ent.label_ == 'SPORTS_NAME':
            sports_name = ent.text
        elif ent.label_ == 'WINNING_POSITION':
            winning_position = ent.text
        elif ent.label_ == 'ORG_YEAR':
            org_year = ent.text

    extracted_info.append({
        'sports_name': sports_name,
        'winning_position': winning_position,
        'org_year': org_year
    })
```

```python
extracted_info
```

```python
# Create a new DataFrame with the extracted information

result_df = pd.DataFrame(extracted_info)
result_df.insert(0, column="Certificate Description",  value=new_data_df['Certificate_Description'])

result_df.head()
```
