## Q&A

```python
from langchain.prompts import PromptTemplate
from langchain_groq import ChatGroq
from langchain.chains import LLMChain
from langchain.memory import ConversationBufferMemory

# Initialize Groq LLM
groq_api_key = "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"  # Replace with your actual API key
llm = ChatGroq(
    api_key=groq_api_key,
    model="mixtral-8x7b-32768",
    temperature=0,
    max_tokens=None,
    timeout=None,
    max_retries=2,
)

# Custom Prompt
custom_prompt_template_for_chatbot = """
You are a knowledgeable assistant specializing in Data Science and Artificial Intelligence (AI).

Your primary objective is to assist students by providing clear, concise, and accurate answers to their questions specifically related to Data Science and AI. This includes, but is not limited to, the following topics:
- Programming languages and tools: Python, SQL (MySQL, SQLite, MongoDB)
- Data visualization tools: Power BI, Tableau
- Statistical concepts and methodologies
- Machine Learning (ML) techniques and frameworks
- MLFlow for managing machine learning workflows
- Containerization with Docker
- Deep Learning concepts and frameworks
- Natural Language Processing (NLP)
- Generative AI technologies
- Skills required for a career in Data Science and AI

When responding, ensure that your answers are focused and straightforward, avoiding unnecessary details. If users ask complex questions, break down your responses into manageable parts and provide step-by-step explanations when needed.

Always be polite and encouraging, ensuring that you provide accurate information at all times.

Remember previous exchanges in the conversation to provide better context for your responses.

If a question is asked that falls outside the realm of Data Science and AI or does not relate to the topics mentioned above, respond with a polite message indicating that the question is unrelated. For example: "I'm sorry, but that topic is outside the scope of Data Science and AI. I'm unable to provide an answer."

Question: {history} Current question: {question}
"""

prompt = PromptTemplate(
    template=custom_prompt_template_for_chatbot,
    input_variables=["history", "question"],
)

# Initialize memory
memory = ConversationBufferMemory(memory_key="history", input_key="question")

# Create chain with memory
chain = LLMChain(
    llm=llm,
    prompt=prompt,
    memory=memory,
)

# Function to invoke the chain
def invoke_chain_with_memory(question):
    response = chain.run(question)
    return response

# Example Usage
if __name__ == "__main__":
    print("Welcome to the AI Assistant!")
    while True:
        user_question = input("You: ")
        if user_question.lower() in ["exit", "quit"]:
            print("Goodbye!")
            break
        response = invoke_chain_with_memory(user_question)
        print(f"Assistant: {response}")
```

## setting Remainder


```python
from datetime import datetime
import time
import threading

def set_reminder(reminder_time, message):
    """Set a reminder for the specified time in 12-hour format."""
    def reminder():
        while True:
            now = datetime.now().strftime("%I:%M %p")  # Get current time in 12-hour format
            if now == reminder_time:
                print(f"Reminder: {message}")
                break
            time.sleep(1)  # Wait 1 second between checks

    # Start the reminder thread
    threading.Thread(target=reminder).start()
    print(f"Reminder set for {reminder_time} with the message: '{message}'")

# Main Program
print("Welcome to the Reminder System!")
reminder_time = input("Enter the time for the reminder (HH:MM AM/PM, e.g., 02:30 PM): ")
reminder_message = input("Enter the reminder message: ")
set_reminder(reminder_time, reminder_message)
```

## Weather report

```python
import requests

def fetch_weather(location):
    """Fetch the weather details for a given location using Open-Meteo."""
    # Geocoding API to get latitude and longitude for the location
    geocode_url = f"https://geocoding-api.open-meteo.com/v1/search?name={location}"
    
    try:
        # Get latitude and longitude of the location
        geocode_response = requests.get(geocode_url)
        geocode_response.raise_for_status()
        geocode_data = geocode_response.json()

        if "results" not in geocode_data or len(geocode_data["results"]) == 0:
            return "Sorry, I couldn't find the location details for that place."

        # Extract latitude and longitude
        latitude = geocode_data["results"][0]["latitude"]
        longitude = geocode_data["results"][0]["longitude"]
        location_name = geocode_data["results"][0]["name"]

        # Weather API call
        weather_url = f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&current_weather=true"
        weather_response = requests.get(weather_url)
        weather_response.raise_for_status()
        weather_data = weather_response.json()

        # Parse the weather data
        if "current_weather" in weather_data:
            current_weather = weather_data["current_weather"]
            temperature = current_weather["temperature"]
            windspeed = current_weather["windspeed"]
            weather_code = current_weather.get("weathercode", -1)

            # Map weather codes to conditions
            weather_conditions = {
                0: "Clear sky",
                1: "Mainly clear",
                2: "Partly cloudy",
                3: "Overcast",
                45: "Foggy",
                48: "Depositing rime fog",
                51: "Light drizzle",
                53: "Moderate drizzle",
                55: "Dense drizzle",
                61: "Slight rain",
                63: "Moderate rain",
                65: "Heavy rain",
                71: "Slight snow",
                73: "Moderate snow",
                75: "Heavy snow",
                80: "Rain showers",
                81: "Moderate rain showers",
                82: "Heavy rain showers",
                95: "Thunderstorm",
                96: "Thunderstorm with hail",
            }
            weather_description = weather_conditions.get(weather_code, "Unknown conditions")

            # Determine if it’s hot, cold, or moderate based on temperature
            if temperature > 30:
                temperature_description = "hot"
            elif temperature < 15:
                temperature_description = "cold"
            else:
                temperature_description = "moderate"

            return (f"The current weather in {location_name} is {weather_description} with a temperature of "
                    f"{temperature}°C ({temperature_description}) and a windspeed of {windspeed} km/h.")
        else:
            return "Sorry, I couldn't fetch the weather details at this time."

    except requests.exceptions.RequestException as e:
        return f"An error occurred: {e}"

# Test the functionality
if __name__ == "__main__":
    location = input("Enter the location: ")
    weather_report = fetch_weather(location)
    print(weather_report)
```

```python
import yt_dlp
import os
import vlc

def play_music(query):
    """Search and play music from YouTube."""
    try:
        # Search YouTube for the query and get the best match
        ydl_opts = {
            'format': 'bestaudio/best',  # Fetch only audio
            'noplaylist': True,         # Ignore playlists
            'quiet': True,              # Suppress output
        }

        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            # Perform a YouTube search
            result = ydl.extract_info(f"ytsearch:{query}", download=False)['entries'][0]
            video_title = result['title']
            video_url = result['url']

            print(f"Now playing: {video_title}")

            # Play the audio using VLC
            player = vlc.MediaPlayer(video_url)
            player.play()

            # Wait until the user stops the music
            while True:
                command = input("Type 'stop' to stop the music: ").strip().lower()
                if command == "stop":
                    player.stop()
                    print("Music stopped.")
                    break

    except Exception as e:
        print(f"An error occurred: {e}")

# Test the functionality
if __name__ == "__main__":
    song_query = input("What song would you like to play? ")
    play_music(song_query)
```

```python
import requests
import re
import xml.etree.ElementTree as ET
import html

def fetch_news_headlines(topic=None):
    """
    Fetch the top 20 recent news headlines using Google News RSS.
    :param topic: Optional search topic for specific news.
    :return: List of news headlines.
    """
    if topic:
        news_url = f"https://news.google.com/rss/search?q={topic.replace(' ', '+')}"
    else:
        news_url = "https://news.google.com/rss"

    try:
        # Fetch the RSS feed content
        response = requests.get(news_url)
        response.raise_for_status()  # Raise error for bad responses

        # Decode and clean the content
        raw_content = response.content.decode('utf-8', errors='ignore')

        # Remove problematic characters using regex
        cleaned_content = re.sub(r'[^\x20-\x7E]+', '', raw_content)  # Remove non-ASCII characters
        cleaned_content = re.sub(r'&(?!amp;|lt;|gt;|quot;|apos;)', '&amp;', cleaned_content)  # Fix & issues

        # Parse the cleaned XML content
        root = ET.fromstring(cleaned_content)

        # Extract headlines
        headlines = []
        for item in root.findall(".//item")[:20]:  # Limit to top 20 news
            title = item.find("title").text
            headlines.append(html.unescape(title.strip()))
        return headlines

    except ET.ParseError as e:
        return f"XML parsing error: {e}"
    except Exception as e:
        return f"An error occurred while fetching news: {e}"

if __name__ == "__main__":
    print("Welcome to the Topic-Specific News Fetcher!")
    
    while True:
        # Ask user for topic
        topic = input("Enter a topic to search for news (or press Enter for general news): ").strip()
        
        if topic.lower() in ["exit", "quit"]:
            print("Goodbye!")
            break

        news_headlines = fetch_news_headlines(topic)

        if isinstance(news_headlines, str):  # Error occurred
            print(news_headlines)
        else:
            print("\nHere are the top 20 recent news headlines:\n")
            for idx, title in enumerate(news_headlines, start=1):
                print(f"{idx}. {title}")
            print("\nYou can search for another topic or type 'exit' to quit.\n")
```

## TO-DO list

```python
import json

# Initialize an empty to-do list
todo_list = []

# File to save and load tasks
TODO_FILE = "todo_list.json"

def save_tasks():
    """Save the current to-do list to a file."""
    try:
        with open(TODO_FILE, "w") as file:
            json.dump(todo_list, file, indent=4)
        return "To-do list saved successfully."
    except Exception as e:
        return f"An error occurred while saving tasks: {e}"

def load_tasks():
    """Load the to-do list from a file."""
    global todo_list
    try:
        with open(TODO_FILE, "r") as file:
            todo_list = json.load(file)
        return "To-do list loaded successfully."
    except FileNotFoundError:
        return "No saved to-do list found. Starting fresh."
    except Exception as e:
        return f"An error occurred while loading tasks: {e}"

def add_task(task_description):
    """Add a new task to the to-do list."""
    todo_list.append({"description": task_description, "status": "pending"})
    return f"Task '{task_description}' added to your to-do list."

def view_tasks():
    """View all tasks in the to-do list."""
    if not todo_list:
        return "Your to-do list is empty."
    response = "Here are your tasks:\n"
    for i, task in enumerate(todo_list, 1):
        status = "✓" if task["status"] == "completed" else "✗"
        response += f"{i}. {task['description']} [{status}]\n"
    return response

def delete_task(task_number):
    """Delete a task from the to-do list by its number."""
    try:
        task_number = int(task_number) - 1
        if 0 <= task_number < len(todo_list):
            removed_task = todo_list.pop(task_number)
            return f"Task '{removed_task['description']}' removed from your to-do list."
        else:
            return "Invalid task number."
    except ValueError:
        return "Please provide a valid task number."

def mark_task_completed(task_number):
    """Mark a task as completed by its number."""
    try:
        task_number = int(task_number) - 1
        if 0 <= task_number < len(todo_list):
            todo_list[task_number]["status"] = "completed"
            return f"Task '{todo_list[task_number]['description']}' marked as completed."
        else:
            return "Invalid task number."
    except ValueError:
        return "Please provide a valid task number."

def clear_tasks():
    """Clear all tasks from the to-do list."""
    global todo_list
    todo_list = []
    return "All tasks have been cleared from your to-do list."

# Interactive Command Line Interface for Testing
if __name__ == "__main__":
    print("Welcome to the Enhanced To-Do List Manager!")
    print(load_tasks())  # Load tasks at startup

    while True:
        print("\nOptions: [add/view/delete/complete/clear/save/load/exit]")
        command = input("What would you like to do? ").strip().lower()

        if command == "add":
            task = input("Enter the task: ").strip()
            print(add_task(task))
        elif command == "view":
            print(view_tasks())
        elif command == "delete":
            task_number = input("Enter the task number to delete: ").strip()
            print(delete_task(task_number))
        elif command == "complete":
            task_number = input("Enter the task number to mark as completed: ").strip()
            print(mark_task_completed(task_number))
        elif command == "clear":
            print(clear_tasks())
        elif command == "save":
            print(save_tasks())
        elif command == "load":
            print(load_tasks())
        elif command == "exit":
            print(save_tasks())  # Save tasks on exit
            print("Goodbye!")
            break
        else:
            print("Invalid command. Please try again.")
```

## Quizz game

```python
from langchain_groq import ChatGroq
from langchain.schema import SystemMessage, HumanMessage
import random

# Initialize ChatGroq LLM
llm = ChatGroq(api_key="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", model="mixtral-8x7b-32768", temperature=0.9)

def generate_random_question(index):
    """
    Use ChatGroq to generate a unique and accurate quiz question.
    """
    topics = ["science", "history", "geography", "sports", "literature", "technology", "music"]
    topic = random.choice(topics)

    prompt = f"""
    You are an advanced quiz master responsible for generating unique and accurate multiple-choice quiz questions on a specific topic.
    Follow these instructions carefully:

    1. Generate a question that is completely unique, clear, concise, and specific to the topic: '{topic}'.
    2. Provide exactly four answer options, labeled A, B, C, and D.
    3. Ensure each answer option is distinct and plausible. Avoid overly obvious or irrelevant options.
    4. Clearly indicate the correct answer using its label (A, B, C, or D) in the format: "Correct Answer: <Correct Option (e.g., A)>".

    **Format your output EXACTLY as follows:**
    - Line 1: Question {index}: <Question text>
    - Line 2: A) <Option A>
    - Line 3: B) <Option B>
    - Line 4: C) <Option C>
    - Line 5: D) <Option D>
    - Line 6: Correct Answer: <Correct Option (e.g., A)>

    **Additional Requirements:**
    - Do NOT include any extra text such as references, explanations, or descriptions.
    - Do NOT repeat or rephrase the question or answers in additional lines.
    - Ensure the format is strictly followed to avoid parsing issues.

    **Example Output:**
    Question 1: What is the capital of France?
    A) Berlin
    B) Madrid
    C) Paris
    D) Rome
    Correct Answer: C

    Generate the question accordingly.
    """
    
    messages = [
        SystemMessage(content="You are a helpful quiz master."),
        HumanMessage(content=prompt)
    ]

    response = llm.invoke(messages)
    response_content = response.content.strip()

    # Validate the response format
    lines = response_content.split("\n")
    if len(lines) != 6 or not lines[0].startswith(f"Question {index}:") or not all(line[1:3] == ") " for line in lines[1:5]):
        print(f"Malformed response:\n{response_content}")
        raise ValueError("The LLM response does not follow the required format.")

    try:
        # Parse the question and answers
        question = {
            "text": lines[0].replace(f"Question {index}: ", "").strip(),
            "options": {line[0]: line[3:].strip() for line in lines[1:5]},
            "correct": lines[5].replace("Correct Answer: ", "").strip()
        }
        return question
    except Exception as e:
        print(f"Error parsing LLM response:\n{response_content}")
        raise ValueError(f"Failed to parse the LLM response: {e}")



def normalize_input(user_input):
    """
    Normalize user input to handle different formats like "A", "Option A", or "A Neil Armstrong".
    """
    # Split user input into words and uppercase the first part
    words = user_input.upper().split()
    first_word = words[0] if words else ""
    rest_of_input = " ".join(words[1:]) if len(words) > 1 else ""
    return first_word, rest_of_input

def quiz_game():
    """
    Conduct a dynamic quiz game with the user.
    """
    print("Welcome to the Quiz Game!")
    print("Answer the following questions by typing A, B, C, or D.")
    print("Let's get started!\n")

    score = 0
    total_questions = 10
    asked_questions = set()
    max_retries = 5  # Limit retries to avoid infinite loops

    for i in range(1, total_questions + 1):
        retries = 0
        question = None

        # Generate a unique question
        while retries < max_retries:
            question = generate_random_question(i)
            if question["text"] not in asked_questions:
                asked_questions.add(question["text"])
                break
            retries += 1
        
        # If no unique question is found after retries
        if question is None or retries == max_retries:
            print("Could not generate a unique question. Skipping this round.")
            continue

        # Display the question
        print(f"Question {i}: {question['text']}")
        for option, text in question["options"].items():
            print(f"{option}) {text}")

        # Get user's answer
        user_input = input("Your answer: ").strip().upper()
        user_option, user_text = normalize_input(user_input)

        # Validate answer
        correct_option = question["correct"]
        correct_text = question["options"][correct_option]

        if (
            user_option == correct_option
            or user_text.lower() == correct_text.lower()
            or f"{user_option} {user_text}".strip().lower() == f"{correct_option} {correct_text}".lower()
        ):
            print("Correct!\n")
            score += 1
        else:
            print(f"Wrong! The correct answer was {correct_option}) {correct_text}.\n")

    # Show final score
    print(f"Quiz Over! You scored {score}/{total_questions}.")
    if score == total_questions:
        print("Excellent! You nailed it!")
    elif score >= total_questions // 2:
        print("Good job! Keep it up!")
    else:
        print("Better luck next time. Keep practicing!")


# Run the Quiz Game
if __name__ == "__main__":
    quiz_game()
```
