```python
#numerical data plotting
sns.relplot(x = 'total_bill', y = 'tip', data=df) #the relationship between two variables, by default it gives scatterplot, kind='line'
```

```python
sns.scatterplot(x = df.total_bill, y=df.tip, data=df, hue = 'smoker')
```

```python
#as the bill increase, the tip also increases
#highest tip was paid by a smoker
sns.relplot(x = 'total_bill', y = 'tip', data=df, hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data=df, hue = 'size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data=df, style = 'size',)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data=df, style = 'smoker', hue = 'time')
```

```python
#relplot to make a line plot>> we generated a data

time = np.arange(500)
value = np.random.randn(500).cumsum()
df = pd.DataFrame({'time': time, 'value':value})
```

```python
df
```

```python
sns.relplot(x = 'time', y='value', kind = 'line', data = df, sort = True)
```
