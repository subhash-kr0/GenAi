# What is matplotlib ?

For the Python programming language, Matplotlib is a popular 2D charting toolkit. It gives you access to a wide range of tools for using Python to build various kinds of static, animated, and interactive visualizations. Matplotlib is a well-liked option for data visualization, scientific charting, and producing publication-quality visuals because of its user-friendly, adaptable, and highly adjustable design.

Using matplotlib you can draw different graph like below:

- Line Plot
- Bar Chart
- Histogram
- Scatter Plot
- Pie Chart
- Area Plot
- Error Chart

etc.

### Matplotlib Line Plot

- A type of data visualization called a Matplotlib line plot shows data points connected by straight lines. It is used to depict the relationship between two numerical variables or the trend of a single variable across a continuous range.

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,7,8]
y= [10,12,14,16,15,17,19]
```

```python
plt.plot(x,y)
```

```python
# Plot line with x_axis and y_axis and title
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.show()
```

plt.plot() methods has more parameters so let's play with some of them

```python
plt.plot(x,y,color='c')
plt.title('Line Plot')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.show()
```

```python
plt.plot(x,y,color='m',marker='o')
plt.title('Line Plot')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.show()
```

```python
plt.plot(x,y,color='g',marker='o',linestyle='--',linewidth=2,markersize=7)
plt.title('Line Plot')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.show()
```

```python
plt.plot(x,y,color='k',marker='o',linestyle=':',linewidth=2,markersize=7)
plt.title('Line Plot')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [8, 5, 9, 11, 7]
y2 = [9, 7, 5, 6, 10]

# Create a plot with multiple lines
plt.plot(x, y1, marker='o', linestyle=':', color='b', label='Line 1')
plt.plot(x, y2, marker='s', linestyle='--', color='r', label='Line 2')

# Add labels, title, and legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines on One Plot")
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Seaborn for data loading and better styling
import pandas as pd

# Load the Iris dataset from Seaborn
iris = sns.load_dataset("iris")

# Calculate the average sepal length for each species
average_sepal_length = iris.groupby("species")["sepal_length"].mean()

# Convert the result to a Pandas DataFrame for easier plotting
average_sepal_length_df = pd.DataFrame(average_sepal_length).reset_index()

# Create the bar plot
plt.figure(figsize=(8, 6))  # Adjust the figure size
plt.bar(average_sepal_length_df["species"], average_sepal_length_df["sepal_length"], color='y')
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")
plt.title("Average Sepal Length for Each Species")
plt.show()
```

### Histogram :  

- A histogram is a type of data visualization that shows how a continuous or numerical dataset is distributed. It counts the number of data points that fall into each bin after dividing the range of values into intervals known as "bins". The resulting graphic sheds light on the data's fundamental distribution by exhibiting characteristics like peaks, valleys, skewness, and gaps.


- Histograms can be easily made with Matplotlib's plt.hist() function.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the histogram
data = np.random.normal(0, 1, 1000)

# Create a basic histogram
plt.hist(data, bins=20, color='y', edgecolor='g')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for two distributions
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(3, 1, 1000)

# Create a stacked histogram
plt.hist([data1, data2], bins=20, color=['skyblue', 'salmon'], edgecolor='black', label=['Distribution 1', 'Distribution 2'])
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Stacked Histogram")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the histogram
data = np.random.gamma(2, 1, 400)

# Create a customized histogram with additional options
plt.hist(data, bins=20, color='green', edgecolor='black', alpha=0.8, density=True)
plt.axvline(data.mean(), color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Customized Histogram")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Seaborn for data loading and better styling

# Load the Iris dataset from Seaborn
iris = sns.load_dataset("iris")

# Create a histogram of sepal lengths
plt.hist(iris["sepal_length"], bins=10, color='#3D3C3A', edgecolor='white')
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")
plt.show()
```

### Scatter Plot  

- A scatter plot is a two-dimensional data visualization that shows distinct data points as dots or markers. The scatter plot's dots are arranged according to their values along the X and Y axes, and each one represents a single data observation. In order to visualize correlations between two numerical variables and spot patterns, clusters, or trends in the data, scatter plots are extremely helpful.


- The plt.scatter() tool provided by Matplotlib makes it simple to build scatter plots.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1,2,3,4,5,7,8,9]
y = [12,15,18,12,9,15,19,12]

# Create a scatter plot
plt.scatter(x, y, color='y', marker='o')

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Scatter Plot")

# Display the plot
plt.show()
```

### Pie Chart

- A pie chart is a circular data visualization used to show how categorical data is distributed like slices of a pie. Each slice represents a category or group, and the size of each slice is proportionate to the percentage or piece of the entire that it represents. Pie charts are particularly effective for displaying portions of a larger whole or outlining the make-up of a dataset.


- Using the plt.pie() function, Matplotlib offers a straightforward method for making pie charts.

```python
import matplotlib.pyplot as plt
```

```python
class_name = ["Python",'Statistic','Machine Learning', 'Deep Learning',
           'MLOPS']
class_size = [75, 85, 92, 90, 87]
```

```python
plt.pie(class_size, labels = classe_name)
plt.show()
```

```python
plt.pie(class_size, labels = classe_name,autopct = "%0.2f%%")
plt.show()
```

```python
explode = [0.02,0,0.1,0,0.03] # To slice the perticuler section

plt.pie(class_size, labels = classe_name, explode = explode, autopct = "%0.2f%%",shadow = True,radius = 1.4,
        startangle = 270)

plt.show()
```

### Area Plot

- An area plot, also called a stacked area plot, is a type of data visualization that shows how various quantitative factors have changed over time while highlighting how they all work together to form the whole. It's particularly good for illustrating how elements of a whole change over time or along a continuous dimension.


- A layered representation of the data is produced in an area plot by layering many variables on top of one another. The cumulative influence of all levels provides insight into the overall trend. Each layer represents a separate category or variable. To distinguish between categories, the space beneath each layer is frequently colored.

```python
import matplotlib.pyplot as plt

# Sample data
years = [2021, 2018, 2022, 2023, 2017]
category1 = [20, 25, 30, 28, 35]
category2 = [10, 15, 20, 18, 25]
category3 = [5, 8, 10, 9, 15]

# Create a area plot
plt.stackplot(years, category1, category2, category3, labels=['Category 1', 'Category 2', 'Category 3'],
              colors=['skyblue', 'lightgreen', 'orange'])

# Add labels and title
plt.xlabel("Years")
plt.ylabel("Values")
plt.title("Area Plot")
plt.legend()

plt.show()
```

# 3D Plots

```python

import numpy as np
import matplotlib.pyplot as plt

ax = plt.figure().add_subplot(projection='3d')

# Plot a sin curve using the x and y axes.
x = np.linspace(0, 1, 100)
y = np.sin(x * 2 * np.pi) / 2 + 0.5
ax.plot(x, y, zs=0, zdir='z', label='curve in (x, y)')

# Plot scatterplot data (20 2D points per colour) on the x and z axes.
colors = ('r', 'g', 'b', 'k')

# Fixing random state for reproducibility
np.random.seed(19680801)

x = np.random.sample(20 * len(colors))
y = np.random.sample(20 * len(colors))
c_list = []
for c in colors:
    c_list.extend([c] * 20)
# By using zdir='y', the y value of these points is fixed to the zs value 0
# and the (x, y) points are plotted on the x and z axes.
ax.scatter(x, y, zs=0, zdir='y', c=c_list, label='points in (x, z)')

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Customize the view angle so it's easier to see that the scatter points lie
# on the plane y=0
ax.view_init(elev=20., azim=-35, roll=0)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# set up the figure and axes
fig = plt.figure(figsize=(9, 3))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

# fake data
_x = np.arange(8)
_y = np.arange(9)
_xx, _yy = np.meshgrid(_x, _y)
x, y = _xx.ravel(), _yy.ravel()

top = x + y
bottom = np.zeros_like(top)
width = depth = 1

ax1.bar3d(x, y, bottom, width, depth, top, shade=True)
ax1.set_title('Shaded')

ax2.bar3d(x, y, bottom, width, depth, top, shade=False)
ax2.set_title('Not Shaded')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define dimensions
Nx, Ny, Nz = 100, 300, 500
X, Y, Z = np.meshgrid(np.arange(Nx), np.arange(Ny), -np.arange(Nz))

# Create fake data
data = (((X+100)**2 + (Y-20)**2 + 2*Z)/1000+1)

kw = {
    'vmin': data.min(),
    'vmax': data.max(),
    'levels': np.linspace(data.min(), data.max(), 10),
}

# Create a figure with 3D ax
fig = plt.figure(figsize=(5, 4))
ax = fig.add_subplot(111, projection='3d')

# Plot contour surfaces
_ = ax.contourf(
    X[:, :, 0], Y[:, :, 0], data[:, :, 0],
    zdir='z', offset=0, **kw
)
_ = ax.contourf(
    X[0, :, :], data[0, :, :], Z[0, :, :],
    zdir='y', offset=0, **kw
)
C = ax.contourf(
    data[:, -1, :], Y[:, -1, :], Z[:, -1, :],
    zdir='x', offset=X.max(), **kw
)
# --


# Set limits of the plot from coord limits
xmin, xmax = X.min(), X.max()
ymin, ymax = Y.min(), Y.max()
zmin, zmax = Z.min(), Z.max()
ax.set(xlim=[xmin, xmax], ylim=[ymin, ymax], zlim=[zmin, zmax])

# Plot edges
edges_kw = dict(color='0.4', linewidth=1, zorder=1e3)
ax.plot([xmax, xmax], [ymin, ymax], 0, **edges_kw)
ax.plot([xmin, xmax], [ymin, ymin], 0, **edges_kw)
ax.plot([xmax, xmax], [ymin, ymin], [zmin, zmax], **edges_kw)

# Set labels and zticks
ax.set(
    xlabel='X [km]',
    ylabel='Y [km]',
    zlabel='Z [m]',
    zticks=[0, -150, -300, -450],
)

# Set zoom and angle view
ax.view_init(40, -30, 0)
ax.set_box_aspect(None, zoom=0.9)

# Colorbar
fig.colorbar(C, ax=ax, fraction=0.02, pad=0.1, label='Name [units]')

# Show Figure
plt.show()
```
