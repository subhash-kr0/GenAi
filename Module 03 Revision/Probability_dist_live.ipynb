```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#discrete distribution>>pmf
#bernoulli distribution
s = np.random.binomial(n=1, p=0.5, size=1000)
h = plt.hist(s, density=True)
plt.title("The bernoulli dist")
plt.show()
```

```python
#binomail dist
n =100 
p=0.1
sample = np.random.binomial(n, p, 10000)
hist = plt.hist(sample,bins=20, density = True)
plt.title("Binomial dis")
plt.show()
```

```python
#poisson distr

sample = np.random.poisson(lam=5, size= 10000)
hist = plt.hist(sample,bins=20)
plt.title("Poisson dis")
plt.show()
```

```python
#continuous distribution>> Normal/gaussian distribution
mu = 0
sigma = 0.1
sample = np.random.normal(mu, sigma, size = 1000)
sns.displot(sample, bins = 30, kde = True)
plt.show()
```

```python
#standard distribution
mu = 0
sigma = 1
sample = np.random.normal(mu, sigma, size = 1000)
sns.displot(sample, bins = 30, kde = True)
plt.show()
```

```python
#uniform distribution
sample = np.random.uniform(-1, 0, 1000)
plt.hist(sample, bins = 15, density=True)
plt.show()
```

```python
#log normal distribution
sample = np.random.lognormal(mean = 0, sigma = 0.5, size = 10000)
sample = sample[np.where(abs(sample) < 5)]
plt.hist(sample, bins=100, density = True)
plt.show()
```

```python
sample = np.random.lognormal(mean = 0, sigma = 0.5, size = 10000)
sample = sample[np.where(abs(sample) < 5)]
plt.hist(np.log(sample), bins=100, density = True)
plt.show()
```

```python
#chisquare implementation
sample = np.random.chisquare(df=1, size = 1000)
# sample = sample[np.where(abs(sample) < 50)]
plt.hist(sample, bins = 100, density = True)
plt.show()
```

```python
#chisquare implementation
sample = np.random.chisquare(df=1, size = 1000)
# sample = sample[np.where(abs(sample) < 50)]
plt.hist(sample, bins = 100, density = True)
plt.show()
```

```python
#T-dist

sample = np.random.standard_t(df=1, size = 10000)
sample = sample[np.where(abs(sample) < 5)]
plt.hist(sample, bins = 100, density = True)
plt.show()
```

```python
#f distribution

sample = np.random.f(dfnum=1, dfden=1,size = 10000)
sample = sample[np.where(abs(sample) < 5)]
plt.hist(sample, bins = 100, density = True)
plt.show()
```
