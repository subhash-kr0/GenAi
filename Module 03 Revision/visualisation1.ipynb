```python
x = [1, 2, 3, 4, 5]
y1 = [5, 1, 7, 2, 8]
y2 = [7, 8, 3, 5, 1]

plt.plot(x, y1, color = 'r', marker = "o", linestyle = '-.', linewidth = 3, markersize = 6, label = 'line y1')
plt.plot(x, y2, color = 'b', marker = "o", linestyle = '-.', linewidth = 3, markersize = 6, label = 'line y2')

plt.xlabel("x points")
plt.ylabel("y points")

plt.title("The line chart")
plt.legend()

plt.grid()
plt.show()
```

```python
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title("sin")

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.title("cos")

plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x))
plt.title("tan")

plt.subplot(2, 2, 4)
plt.plot(x, np.cos(x+1))

plt.title("cos x+1")
plt.show()
```

```python
langs = ['C', 'Java', 'Python']
students = [20, 40, 110]

plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
explode = (0.0, 0.1, 0.2)
langs = ['C', 'Java', 'Python']
students = [20, 40, 110]

plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = explode, shadow = True)
plt.title("Pie chart")
plt.show()
```
