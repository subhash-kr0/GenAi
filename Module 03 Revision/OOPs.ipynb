# N.B.: parent class will not have access to the property of child class but child class will have access to both

```python
class Fruit():
  def fruit_info(self):
    print("inside parent class")

class Apple(Fruit):
  def apple_info(self):
    print("inside child class")
```

```python
obj = Apple()
```

```python
obj.apple_info()
```

```python
obj.fruit_info()
```

```python
class Fruit():    # Parent Class
  def fruit_info(self):
    print("inside parent class")

class Apple(Fruit):   # child Class
  def fruit_info(self):
    print("inside child class")

  def apple_info(self):
    print("inside child class")
```

```python
apple = Apple()
```

```python
apple.apple_info()
```

```python
 # Multi-level inheritence

 class Grandfather():
  def grandfather_prop(self):
    print(" I am grand father")

 class Father(Grandfather):
  def father_prop(self):
    print(" I am father, having proprty from grandfather and father")

 class Son(Father):
  def son_prop(self):
    print(" I am son, having prop from grandfather and father")
```

```python
son = Son()
```

```python
son.father_prop()
```

```python
son.grandfather_prop()
```

```python
class Parentclass1():
  def method1(self):
    print(" method1 of parentclass1")

class Parentclass2():
  def method2(self):
    print(" method2 of parentclass2")
```

```python
class Childclass(Parentclass1, Parentclass2):
  def child_method(self):
    print(" child method")

child = Childclass()
```

```python
child.method1()
```

```python
child.method2()
```

```python
# interview kind of concept
class A:
  def method(Self):
    print("method of class A")

class B(A):
  def method(Self):
    print("method of class B")

class C(A):
  def method(Self):
    print("method of class C")

class D(B,C):
  pass

class D(C,B):
  pass
```

```python
d = D()
```

```python
d.method()   # Diamond problem


      A

B           C

      D

# Method Resolution Order (MRO) algorithm called C3 linerization
```

# Abstraction
  
### -> concept of hiding the complex implementation

```python
import abc
```

```python
class PWSkills():

  @abc.abstractmethod
  def student_details(self):
    pass

  @abc.abstractmethod
  def student_assignment(self):
    pass

  @abc.abstractmethod
  def student_marks(self):
    pass
```

```python
class DataScience(PWSkills):
  def student_details(self):
    return " data science course student details"
  def student_marks(self):
    return "give data science course student marks"
```

```python
class WebDev(PWSkills):
  def student_assignment(self):
    return " This will give student assignment of web dev class"
  def student_marks(self):
    return "give web development course student marks"
```

```python
ds = DataScience()
```

```python
ds.student_marks()
```

```python
wb = WebDev()
```

```python
wb.student_marks()
```

```python
class Shape:
  @abc.abstractmethod
  def calculate_area(self):
    pass

class Rectangle(Shape):
  def calculate_area(self):
    return "Area of rectangle: length * breadth"

class Circle(Shape):
  def calculate_area(self):
    return "Area of a circle: pi r**2"
```

```python
rect = Rectangle()
rect.calculate_area()
```

```python
c = Circle()
c.calculate_area()
```

```python
class_obj = [obj1, obj2]
```
