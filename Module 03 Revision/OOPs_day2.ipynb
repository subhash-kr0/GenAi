# Method overriding :

In Python, method overriding is the process of providing a different implementation for a method that is already defined in the superclass within a subclass

```python
class Animal:
  def sound(self):
    print(" animal sound")

class Cat(Animal):
  def sound(self):
    print(" cat meows")
```

```python
cat = Cat()
```

```python
cat.sound()
```

### In Method Overriding >> child class is powerful than parent class

# duck typing in polymorphosm

```python
class Duck:
  def quack(self):
    print("Quack !")

  def fly(self):
    print(" Duck flying")

class Dog:
  def bark(self):
    print(" Woof !")

  def fly(self):
    print(" Dog cannot fly")

def animal_fly(animal):
  animal.fly()

# creat instances from duck
duck =  Duck()

# creat instances from dog
dog =  Dog()

animal_fly(duck)
animal_fly(dog)
```

```python
class Student:
  def __init__(self, name, degree):
    self.name = name
    self.__degree = degree

  def show(self):
    #accessing both public and private data member
    print("name", self.name, 'degree  - private', self.__degree)
```

```python
class Car:
  def __init__(self, year, make, speed, model):
    self.__year = year
    self.__make = make
    self.__speed = speed
    self.__model = model

  def set_speed(self, speed):
    self.__speed =  0 if speed <0 else speed

  def get_speed(self):
    return self.__speed
```

```python
car = Car("1995", "Maruti", "80", "Brezza")
```

```python
class Bank:

  def __init__(self, balance):
    self.__balance = balance   # private variable, cannot be accessed from outside

  def deposit(self, amount):
    self.__balance =  self.__balance + amount

  def withdraw(self, amount):
    if self.__balance >= amount:
      self.__balance = self.__balance - amount
      return True   # successfully withdrawn
    else:
      return False

  def get_balance(self):
    return self.__balance
```

```python
acc = Bank(1000)
```

```python
acc.get_balance()
```

```python
acc.deposit(10000)
```

```python
acc.get_balance()
```

```python
acc.withdraw(10500)
```

```python
acc.get_balance()
```

```python
acc.withdraw(1000)
```
