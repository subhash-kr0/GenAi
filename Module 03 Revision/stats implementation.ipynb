```python
#clt>> the smaple mean distribution willm be approximated to normal dist


population = np.random.binomial(10, 0.5, 100000)
population
```

```python
sns.distplot(population)
```

```python
sample_size = 3000
mean_of_samples = []
for i in range(1, 1000):
    sample = np.random.choice(population, size = sample_size)
    mean_of_samples.append(np.mean(sample))
len(mean_of_samples)
```

```python
sns.distplot(mean_of_samples, kde = True, color = "red")
```
