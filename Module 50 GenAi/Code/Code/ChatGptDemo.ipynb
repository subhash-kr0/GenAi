```python
import numpy as np
from collections import Counter

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def k_nearest_neighbors(X_train, y_train, x_test, k):
    distances = []

    # Calculate the Euclidean distance from x_test to all training samples
    for i in range(len(X_train)):
        dist = euclidean_distance(x_test, X_train[i])
        distances.append((dist, y_train[i]))

    # Sort distances in ascending order and get the k nearest labels
    distances.sort(key=lambda x: x[0])
    k_nearest_labels = [label for (_, label) in distances[:k]]

    # Determine the majority class among the k nearest neighbors
    majority_class = Counter(k_nearest_labels).most_common(1)[0][0]

    return majority_class

# ------------- Visualization -------------

class_0 = X_train[y_train == 0]
class_1 = X_train[y_train == 1]

plt.scatter(class_0[:, 0], class_0[:, 1], color='blue', label='Class 0')
plt.scatter(class_1[:, 0], class_1[:, 1], color='red', label='Class 1')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter Plot of KNN")
plt.legend()

plt.show()




# Example usage:
X_train = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])
y_train = np.array([0, 1, 0, 1])
x_test = np.array([3, 3])
k = 3

result = k_nearest_neighbors(X_train, y_train, x_test, k)
print(f"The predicted class for the test instance is: {result}")
```
