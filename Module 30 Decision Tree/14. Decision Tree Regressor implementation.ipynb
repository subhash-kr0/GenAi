```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import fetch_california_housing
data=fetch_california_housing()
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df['Price']=data.target
```

```python
df
```

```python
df.shape
```

```python
#taking 20 percent of the data for our case
df = df.sample(frac = 0.20)
```

```python
df.shape
```

```python
df
```

```python
#independent and dependent features
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred, y_test)
```

```python
#hyperparameter tuning

parameter={
 'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5,6],
  'max_features':['auto', 'sqrt', 'log2']
    
}
regressor=DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
model = GridSearchCV(regressor, param_grid=parameter, cv = 3, scoring = 'neg_mean_squared_error')
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
final_model = DecisionTreeRegressor(criterion = 'poisson', max_depth = 6, max_features = 'auto', splitter = 'best')
```

```python
final_model.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (12, 10))
tree.plot_tree(final_model, filled = True)
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_pred, y_test)
```
